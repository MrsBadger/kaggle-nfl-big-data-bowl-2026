```python
param_grid = {"estimator__min_samples_split": [2, 5, 10, 20],
    "estimator__min_samples_leaf": [1, 2, 4, 8],
    "estimator__max_features": ["sqrt", "log2", None],
    "estimator__criterion": ["squared_error","absolute_error","poisson"],}



groups = df.data["game_id"]

gkf = GroupKFold(n_splits=5)
X = df.preprocessed[df.feature_cols].copy()
y = df.preprocessed[["target_dx", "target_dy"]]


model1 = MultiOutputRegressor(RandomForestRegressor())

rs=RandomizedSearchCV(model1,param_grid,scoring='neg_mean_squared_error',cv=gkf,n_iter=10, verbose=1)

rs.fit(X, y, groups=groups)
```
